```python
from scipy.stats import hypergeom
from scipy.stats import binom
from decimal import *
import numpy as np
import os
import math
from cycler import cycler
%matplotlib inline  
#import matplotlib as mpl
#mpl.use('Agg')
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.ticker import FormatStrFormatter
np.set_printoptions(threshold=np.nan)
```

```python
def plot_cummulative_over_V(Vm,N,f):
    O=N*f
    
    p1=[]
    p2=[]
    p3=[]
    p4=[]
    for Vi in Vm:
        p_i = math.floor(Vi/2)
        p1.append(1-hypergeom.cdf(p_i, N, O, Vi))
        p2.append(hypergeom.sf(p_i, N, O, Vi))
        p3.append(1-binom.cdf(p_i,Vi,f))
        p4.append(binom.sf(p_i,Vi,f))
    
    return (p1,p2,p3,p4)
   
```

```python
def plot_cummulative_over_O(V,Om,f):
    
    
    p1=[]
    p2=[]
    p3=[]
    p4=[]
    p_i = math.floor(V/2)
    for Oi in Om:
        Ni=Oi*(1/f)
        p1.append(1-hypergeom.cdf(p_i, Ni, Oi, V))
        p2.append(hypergeom.sf(p_i, Ni, Oi, V))
        p3.append(1-binom.cdf(p_i,V,f))
        p4.append(binom.sf(p_i,V,f))
    
    return (p1,p2,p3,p4)
   
```

```python
Vm = np.arange(100,5000, 50)


(p1,p2,p3,p4) = plot_cummulative_over_V(Vm,10000,0.3)
plt.plot(Vm,p1, label='hypergeom cdf')
plt.plot(Vm, p2, label = 'hypergeom sf')
plt.plot(Vm, p3, label = 'binom cdf')
plt.plot(Vm,p4, label = 'binom sf')
plt.yscale('log')
plt.legend(loc='upper right')
```

```python
Om = np.arange(10000,50000, 50)
(p1,p2,p3,p4) = plot_cummulative_over_O(200,Om,0.3)
plt.plot(Om,p1, label='hypergeom cdf')
plt.plot(Om, p2, label = 'hypergeom sf')
plt.plot(Om, p3, label = 'binom cdf')
plt.plot(Om,p4, label = 'binom sf')
plt.yscale('log')
plt.legend(loc='upper right')
```
